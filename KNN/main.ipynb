```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
```

```python
df = pd.read_csv('./data/car.data')

df.head()
```

```python
df.info()
```

```python
columns = df.columns

label_encoder = LabelEncoder()

for column in  columns:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

```python
features = ['buying','maint','door','persons','lug_boot','safety']
predict = 'class'

x = df.drop(predict,axis=1)
y = df[predict]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.1)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)

knn.fit(X_train,Y_train)

accuracy = knn.score(X_test,Y_test)

print(f'Accuracy: {accuracy}')
```

```python
predicted_class = knn.predict(X_test)
classes_by_encoding = ['unacc','acc','good','vgood']
for i in range(len(predicted_class)):
    print(f'Predicted value: {classes_by_encoding[predicted_class[i]]}, Actual value: {classes_by_encoding[Y_test.iloc[i]]}')
```
